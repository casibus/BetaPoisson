```python
from betapoisson import BetaPoisson
import numpy as np
from matplotlib import pyplot as plt
```

```python
N_repetitions = 1000
N_samples = 10
displacements = np.random.randn(N_repetitions*2).reshape((N_repetitions, 2))*3
repetitions = [np.random.randn(N_samples*2).reshape((N_samples, 2)) + displacements[i] for i in range(len(displacements))]
repetitions = np.array(repetitions)
print(repetitions.shape)
plt.figure()
plt.scatter(repetitions.T[0], repetitions.T[1])
plt.show()

h = np.array([np.histogram2d(r.T[0], r.T[1], range=[[-5,5], [-5,5]], bins = 5)[0] for r in repetitions])

for h_ in h[:10]:
    plt.matshow(h_)
```

```python
plt.hist(np.array(h)[:,2,2],range=[-.5,10.5], bins = 10)
```

```python
counts_at_center = np.array(h)[:,2,2]
bp = BetaPoisson()
bp.y = counts_at_center
res = bp.fit(counts_at_center)
```
